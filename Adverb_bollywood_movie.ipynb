```python
import pandas as pd
from textblob import TextBlob
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the datasets for female and male adjectives and verbs
female_adjverb_df = pd.read_csv('/content/female_adjverb_modif.csv',sep='\t',header=None, names=['verbs', 'adjectives'])
male_adjverb_df = pd.read_csv('/content/male_adjverb_modif.csv',sep='\t',header=None, names=['verbs', 'adjectives'])
```

```python
def get_sentiment(text):
    if not text:  # Handle empty text cases
        return 0
    blob = TextBlob(text)
    return blob.sentiment.polarity  # Polarity score (-1 to 1)

# Apply sentiment analysis to 'verbs' and 'adjectives' columns
female_adjverb_df['verb_sentiment'] = female_adjverb_df['verbs'].apply(get_sentiment)
female_adjverb_df['adj_sentiment'] = female_adjverb_df['adjectives'].apply(get_sentiment)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate average sentiment for verbs and adjectives
average_verb_sentiment = female_adjverb_df['verb_sentiment'].mean()
average_adj_sentiment = female_adjverb_df['adj_sentiment'].mean()

print(f"Average female Sentiment for Verbs: {average_verb_sentiment}")
print(f"Average female Sentiment for Adjectives: {average_adj_sentiment}")

# Plot the sentiment distribution for both verbs and adjectives
plt.figure(figsize=(12, 6))
sns.histplot(female_adjverb_df['verb_sentiment'], kde=True, color='blue', label='Verbs Sentiment', bins=20)
sns.histplot(female_adjverb_df['adj_sentiment'], kde=True, color='red', label='Adjectives Sentiment', bins=20)
plt.legend()
plt.title('Sentiment Distribution for Verbs and Adjectives (Female)')
plt.xlabel('Sentiment Polarity')
plt.ylabel('Frequency')
plt.show()
```

```python
male_adjverb_df['verb_sentiment'] = male_adjverb_df['verbs'].apply(get_sentiment)
male_adjverb_df['adj_sentiment'] = male_adjverb_df['adjectives'].apply(get_sentiment)
```

```python
# Calculate average sentiment for verbs and adjectives
average_verb_sentiment = male_adjverb_df['verb_sentiment'].mean()
average_adj_sentiment = male_adjverb_df['adj_sentiment'].mean()

print(f"Average male Sentiment for Verbs: {average_verb_sentiment}")
print(f"Average male Sentiment for Adjectives: {average_adj_sentiment}")

# Plot the sentiment distribution for both verbs and adjectives
plt.figure(figsize=(12, 6))
sns.histplot(male_adjverb_df['verb_sentiment'], kde=True, color='blue', label='Verbs Sentiment', bins=20)
sns.histplot(male_adjverb_df['adj_sentiment'], kde=True, color='red', label='Adjectives Sentiment', bins=20)
plt.legend()
plt.title('Sentiment Distribution for Verbs and Adjectives (Female)')
plt.xlabel('Sentiment Polarity')
plt.ylabel('Frequency')
plt.show()
```
